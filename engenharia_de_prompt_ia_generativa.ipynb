Neste curso, vou construir arquiteturas RAG com LLMs utilizando LangChain
, explorando embeddings, busca semântica, vector stores como FAISS e ChromaDB, além da criação de agentes conversacionais inteligentes. Também irei aplicar técnicas de recuperação, processamento de documentos e re-rank para gerar respostas mais relevantes e contextualizadas com IA generativa.
